# z_test

```python
import scipy.stats as st
import scipy.special as sst
```

```python
def z_table(alpha,yon):
    if yon == 'buyuktur':
        z_table = st.norm.ppf(1-alpha)
        return [z_table]
    if yon == 'kucuktur':
        z_table = st.norm.ppf(alpha)
        return [z_table]
    else:
        alpha_1 = alpha/2.0
        z_table_1 = st.norm.ppf(alpha_1)
        z_table_2 = -z_table_1
        
        return [z_table_1, z_table_2]  
```

```python
a= st.t.ppf(0.975,16)
a1= sst.stdtrit(16,0.975)
print(a,a1)
```

```python
def z_table_yorum(z_hesap,z_table,yon):
    if yon == 'buyuktur':
        if z_hesap > z_table[0]:
            print('Ho red')
        else:
            print('Ho kabul')
    elif yon == 'kucuktur':
        if z_hesap < z_table[0]:
            print('Ho red')
        else:
            print('Ho kabul')
    else:
        if z_hesap < z_table[0] or z_hesap > z_table[1]:
            print('Ho red')
        else:
            print('Ho kabul')
```

```python
def hesapla():
    x_ortalama = float(x_ortalama_entry.get())
    mü = float(mü_entry.get())
    sigma = float(sigma_entry.get())
    alpha = float(alpha_entry.get())
    n = float(n_entry.get())
    yon = str(yon_entry.get())
    
    z_degeri = (x_ortalama-mü)/(sigma/(n**0.5))
    
    if yon == 'buyuktur':
        p_value = st.norm.sf(abs(z_degeri))
        z_table = st.norm.ppf(1-alpha)
        z_table_degeri = [z_table]
        if z_degeri > z_table_degeri[0]:
            z_yorumu = 'Ho red'
        else:
            z_yorumu = 'Ho kabul'
          
    elif yon == 'kucuktur':
        p_value = st.norm.sf(abs(z_degeri))
        z_table = st.norm.ppf(alpha)
        z_table_degeri = [z_table]
        if z_degeri < z_table_degeri[0]:
            z_yorumu = 'Ho red'
        else:
            z_yorumu = 'Ho kabul'
        
        
    elif yon == 'esittir':
        p_value = st.norm.sf(abs(z_degeri))*2
        alpha_1 = alpha/2.0
        z_table_1 = st.norm.ppf(alpha_1)
        z_table_2 = -z_table_1
        
        z_table_degeri = [z_table_1, z_table_2] 
        if z_degeri < z_table_degeri[0] or z_degeri > z_table_degeri[1]:
            z_yorumu  = 'Ho red' 
        else:
            z_yorumu = 'Ho kabul'
            
    sonuc = [z_degeri, z_table_degeri, z_yorumu, p_value]
    
    mesaj_1 = 'z degeri : {}'.format(sonuc[0])
    mesaj_1_label = Label(pencere, text = mesaj_1,justify='left',anchor='w', bg='#0A8E85')
    mesaj_1_label.place(x=200, y=10)
    
    mesaj_2 = 'z tablo degeri : {}'.format(sonuc[1])
    mesaj_2_label = Label(pencere, text = mesaj_2,justify='left',anchor='w', bg='#0A8E85')
    mesaj_2_label.place(x=200, y=40)
    
    mesaj_3 = 'Sonuc : {}'.format(sonuc[2])
    mesaj_3_label = Label(pencere, text = mesaj_3,justify='left',anchor='w', bg='#0A8E85')
    mesaj_3_label.place(x=200, y=70)

    mesaj_4 = 'P Degeri : {}'.format(sonuc[3])
    mesaj_4_label = Label(pencere, text = mesaj_4,justify='left',anchor='w', bg='#0A8E85')
    mesaj_4_label.place(x=200, y=100)
```
